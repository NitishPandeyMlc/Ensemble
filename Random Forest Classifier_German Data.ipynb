```python
import pandas as pd
import numpy as np
```

```python
gdf=pd.read_csv('german_credit.csv')
```

```python
gdf.columns
```

```python
gdf.head()
```

```python
for feature in gdf.columns:
    if gdf[feature].dtype =='object':
           gdf[feature] =pd.Categorical(gdf[feature]).codes
```

```python
gdf.info()
```

```python
X=gdf.drop('default',axis=1)
```

```python
y=gdf['default']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfst=RandomForestClassifier(n_estimators=8)
```

```python
rfst.fit(X_train,y_train)
```

```python
rfst.score(X_train,y_train)
```

```python
rfst.score(X_test,y_test)
```
